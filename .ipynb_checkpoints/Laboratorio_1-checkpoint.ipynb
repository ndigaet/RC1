# Laboratorio 1 - Simulación Monty Hall

```python
import numpy as np
```

```python
N = 1000

lista_sincambio = []
lista_cambioazar = []
lista_cambio = []

total_sincambio=0
total_azar=0
total_cambio=0

#repito N veces el experimento
for i in range(N):
    #pongo cabras (0) en todas las puertas e inicializo variables
    puertas = [0, 0, 0]
    puertacabra = -1
    eleccion_azar = -1
    aleccion_cambio = -1
    
    #tiro 1 dado para saber en qué puerta va a estar el auto
    puertaauto = np.random.randint(3)
    #pongo el auto (1) en la puerta 
    puertas[puertaauto] = 1
    
    #tiro 1 dado para que el jugador elija una puerta    
    eleccion = np.random.randint(3)    
    
    #evaluo si la puerta elegida tiene el auto
    if (puertas[eleccion] == 1):
        #elijo al azar alguna de las otras 2 puertas donde hay una cabra
        puertacabra = np.random.randint(3)
        #mientras el dado me de la puerta que tiene el auto, vuelvo a tirar el dado hasta encontrar una cabra
        while (puertas[puertacabra] == 1):
            puertacabra = np.random.randint(3)
        #al salir del while tengo una puerta al azar con una cabra que va a quedar cerrada
    else:
        #elegi una puerta donde está la cabra, la puerta queda cerrada
        puertacabra = eleccion

    #tiro 1 dado para decidir al azar si cambio de puerta, si es 0 no cambia, 1 cambia
    cambia = np.random.randint(2)
    if (cambia == 1):    
        #cambia la puerta de elección azar, si tenía la cabra ahora tengo el auto
        if(eleccion==puertacabra):
            eleccion_azar=puertaauto
        else:
            #había elegido el auto, ahora me toca la cabra
            eleccion_azar=puertacabra
    else:
        #el azar dijo que no cambia la puerta, se queda con la elegida
        eleccion_azar=eleccion

    #empiezo a sumar las chances
    # si no cambia ¿ganó?        
    if (eleccion == puertaauto):
        total_sincambio +=1
    
    # al azar el cambio de puerta ¿ganó?
    if (eleccion_azar == puertaauto):
        total_azar +=1
        
    # si cambia, ¿ganó?
    if(eleccion==puertacabra):
        eleccion_cambio=puertaauto
    else:
        eleccion_cambior=puertacabra
    if(eleccion_cambio == puertaauto):
        total_cambio += 1

    #agrego a las listas el resultado de la simulación
    lista_sincambio.append(total_sincambio)
    lista_cambioazar.append(total_azar)
    lista_cambio.append(total_cambio)
```

```python
total_sincambio, total_cambio, total_azar
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(16,8))
plt.plot(lista_sincambio, label='No cambio')
plt.plot(lista_cambio, label='Cambio')
plt.plot(lista_cambioazar, label='Al azar')
plt.xlabel('Iteración')
plt.ylabel('hits')
plt.legend()
plt.show()
```

```python
N = 1000 # Cantidad de simulaciones

no_change = 0 # Cantidad de veces que me llevo el auto sin cambiar de puerta
change = 0 # Cantidad de veces que me llevo el auto cambiando de puerta
rand_sel = 0 # Cantidad de veces que me llevo el auto eligiendo al azar entre las dos puertas restantes


# Hits acumulados
change_acum = [] 
no_change_acum = []
rand_sel_acum = []

for i in range(N):
    # Ubicación del auto (una de las 3 puertas al azar)
    car_pos = np.random.randint(3)
    # Selecciono una puerta al azar
    selected_door = np.random.randint(3)
    doors = [0,1,2] # Lista con las 3 opciones de puertas
    doors.remove(selected_door) # Me quedo con las dos puertas que no elegí
    if selected_door == car_pos:
        # No cambio de puerta y era la puerta donde estaba el auto
        no_change += 1
        # Abro cualquiera de las dos puertas restantes 
        door_to_open = doors[np.random.randint(2)]
    else:
        # Remuevo la puerta que tiene el auto. No la puedo abrir
        doors.remove(car_pos)
        # Solo me me queda una puerta para abrir
        door_to_open = doors[0]

    selected_door_change = [0, 1, 2] # Inicialmente contiene las 3 puertas pero me tengo que quedar con la puerta que elijo si cambio
    # Remuevo la que abrio y me quedo con las dos posibles
    selected_door_change.remove(door_to_open)

    if selected_door_change[np.random.randint(2)] == car_pos:
        # Elijo al azar entre las dos puertas restantes y coincide con el auto
        rand_sel += 1

    # Remuevo la que elegi primero ya que ahora cambié (cambio de puerta)
    selected_door_change.remove(selected_door)

    if selected_door_change[0] == car_pos:
        # Si coinciden la que elegí cambiando con la del auto
        change += 1
        
    change_acum.append(change)
    no_change_acum.append(no_change)
    rand_sel_acum.append(rand_sel)
```

```python
no_change, change, rand_sel
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(no_change_acum, label='No cambio')
plt.plot(change_acum, label='Cambio')
plt.plot(rand_sel_acum, label='Al azar')
plt.xlabel('Iteración')
plt.ylabel('hits')
plt.legend()
plt.show()
```
